```python
import requests
```

```python
with open('Nency_139-1_dummy.cif', 'r+') as f:
    cif_content = f.read()

with open('Nency_139-2_clean.cif', 'r+') as f:
    cif_content = f.read()
data = {
  "fileContent": cif_content,
  "extension": "cif"
}
```

```python
oxi = requests.post('http://oximachine.materialscloud.io/api/v1/oximachine/', data = {'cifFile': cif_content})
print(oxi)
```

```python
import base64

with open('Nency_139-2_clean.cif', "rb") as cif_file:
    encoded_cif_content = base64.b64encode(cif_file.read()).decode("utf-8")
print(encoded_cif_content)
```

```python
url = "http://localhost:8000/api/v1/all-models/"

res = requests.post(url, json = data)
print(res.json())
```
